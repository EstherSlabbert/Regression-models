import needed packages

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

reads relevant information of age and insurance costs

```python
df = pd.read_csv(r'insurance.csv', usecols= ['age', 'charges'])
df.head()
```

creates a scatterplot of the information above

```python
plt.figure()
sns.scatterplot(data=df, x='age', y='charges')
plt.show()
plt.close()
```

fits data to a model

```python
from sklearn.linear_model import LinearRegression
linear_reg_model = LinearRegression()
X = df['age'].values.reshape(-1, 1)
Y = df['charges'].values
linear_reg_model.fit(X, Y)
y_pred = linear_reg_model.predict(X)
```

uses model above to make prediction(s)

```python
unk_x = [[70]]

x_pred = np.append(X, unk_x).reshape(-1,1)
y_pred = linear_reg_model.predict(x_pred)

plt.scatter(X,Y,color = 'b')
plt.plot(x_pred,y_pred,color = 'r')
plt.xlabel('age')
plt.ylabel('charges')

print("If they are aged 70 they should be charged approximately:", linear_reg_model.predict(unk_x))
```

redefines variables for scatterplot with best fit line

```python
X = df['age'].values.reshape(-1, 1)
Y = df['charges'].values
linear_reg_model.fit(X, Y)
y_pred = linear_reg_model.predict(X)
```

creates a line of best fit scatterplot

```python
plt.scatter(X,Y,color = 'b')
plt.plot(X,y_pred,color = 'r')
plt.ylabel('Charges')
plt.xlabel('Age')
plt.show()

plt.close()
```
